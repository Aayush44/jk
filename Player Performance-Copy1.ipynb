```python
# Feature Engineering 
# Approach: If team1 won the match, label 1 will be assigned, else if team2 won, label 2 will be assigned.
# Converting team1 and team2 from categorical variables to continuous inputs using pandas function pd.get_dummies. 
# This variable has only two answer choices: team1 and team2. 
# It creates a new dataframe which consists of zeros and ones. 
# Here dataframe will have a one depending on the team of a particular game in this case.


MatchAnaly12 = MatchAnaly12.reset_index(drop=True)
MatchAnaly12.loc[MatchAnaly12.Team2 == MatchAnaly12.Players_1,'Team_2']=1
MatchAnaly12.loc[MatchAnaly12.Team2 == MatchAnaly12.Players_2, 'Team_2']=2
MatchAnaly12.loc[MatchAnaly12.Team2 == MatchAnaly12.Players_3, 'Team_2']=3
MatchAnaly12 = MatchAnaly12.drop(['Team_2'], axis=1)
MatchAnaly12.head()
```

```python
#adding ICC rantings
#the team which is positioned higher on the ICC Ranking will be considered "favourite" for the match
#and therefore, will be positioned under the "Team_1" column

# Loading new datasets
strike = pd.read_csv('strike.csv') 
fixtures = pd.read_csv('fixtures2.csv')

# List for storing the group stage games
pred_set = []
```

```python
# Create new columns with ranking position of each team
fixtures.insert(1, 'first_position', fixtures['Players_1'].map(strike.set_index('Player')['Strike']))
fixtures.insert(2, 'second_position', fixtures['Players_2'].map(strike.set_index('Player')['Strike']))
fixtures.insert(2, 'third_position', fixtures['Players_3'].map(strike.set_index('Player')['Strike']))

# We only need the group stage games
#slice the dataset

fixtures = fixtures.iloc[:50, :]
fixtures.tail()
```

```python
# Loop to add teams to new prediction dataset based on the ranking position of each team
for index, row in fixtures.iterrows():
    if row['first_position'] < row['second_position']:
        pred_set.append({'Players_1': row['Players_1'], 'Players_2': row['Players_2'], 'Team2': None})
    else:
        pred_set.append({'Players_1': row['Players_2'], 'Players_2': row['Players_1'], 'Team2': None})
        
pred_set = pd.DataFrame(pred_set)
backup_pred_set = pred_set
pred_set.head()
```

```python
# List of tuples before 
semi = [('C Carter', 'PJ Moor'),('DG Brownlie', 'DM Gondaria')]
```

```python
def clean_and_predict(matches, strike, final, logreg):

# Initialization of auxiliary list for data cleaning
    positions = []

 # Loop to retrieve each team's position according to ICC ranking
    for match in matches:
        positions.append(strike.loc[strike['Player'] == match[0],'Strike'].iloc[0])
        positions.append(strike.loc[strike['Player'] == match[1],'Strike'].iloc[0])
    
    # Creating the DataFrame for prediction
    pred_set = []

    # Initializing iterators for while loop
    i = 0
    j = 0

    # 'i' will be the iterator for the 'positions' list, and 'j' for the list of matches (list of tuples)
    while i < len(positions):
        dict1 = {}

        # If position of first team is better then this team will be the 'Team_1' team, and vice-versa
        if positions[i] < positions[i + 1]:
            dict1.update({'Players_1': matches[j][0], 'Players_2': matches[j][1]})
        else:
            dict1.update({'Players_1': matches[j][1], 'Players_2': matches[j][0]})

        # Append updated dictionary to the list, that will later be converted into a DataFrame
        pred_set.append(dict1)
        i += 2
        j += 1
        
        # Convert list into DataFrame
    pred_set = pd.DataFrame(pred_set)
    backup_pred_set = pred_set

    # Get dummy variables and drop winning_team column
    pred_set = pd.get_dummies(pred_set, prefix=['Players_1', 'Players_2'], columns=['Players_1', 'Players_2'])
    # Add missing columns compared to the model's training dataset
    missing_cols2 = set(final.columns) - set(pred_set.columns)
    for c in missing_cols2:
        pred_set[c] = 0
    pred_set = pred_set[final.columns]

    pred_set = pred_set.drop(['Team2'], axis=1)

    # Predict!
    predictions = logreg.predict(pred_set)
    for i in range(len(pred_set)):
        print(backup_pred_set.iloc[i, 1] + " and " + backup_pred_set.iloc[i, 0])
        if predictions[i] == 1:
            print("Players_1r: " + backup_pred_set.iloc[i, 1])
        else:
            print("Players_1: " + backup_pred_set.iloc[i, 0])
        print("")
        
       
```

```python
clean_and_predict(semi, strike, final, rf)
```
